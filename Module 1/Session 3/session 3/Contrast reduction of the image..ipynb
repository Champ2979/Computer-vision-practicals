## Importing necessary libraries.

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import PIL as pillow
```

```python
image = cv2.imread('high_contrast.jpg')
plt.imshow(image)
```

```python
imagem = cv2.bitwise_not (image) # basic arithemetic bitwise operator. (0s to 1s and 1s to 0s)
plt.imshow(imagem)
cv2.imwrite('output.jpg', imagem)
```
