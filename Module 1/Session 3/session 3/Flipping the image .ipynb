## Importing necessary libraries.

```python
import cv2
from matplotlib import pyplot as plt
```

## Reading the image.

```python
img = cv2.imread('open_cv.jpg')
name = 'OpenCv Python Image'
plt.title (name)
plt.imshow(img)
```

## Flipping the image horizontally 

```python
horizontal_flip = cv2.flip(img, 1)
name = 'OpenCV Python Flip Image Horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

## Flipping the image vertically

#### The following code fragment will be used to flip the image vertically. 

```python
vertical_flip = cv2.flip(img, 0)
name = 'OpenCV Python Flip Image Vertically'
plt.title(name)
plt.imshow(vertical_flip)
```

## Flipping the image Vertically and Horizontally.

#### This code fragment will flip the image vertically and horizontally.

```python
vertical_horizontal_flip = cv2.flip(img, -1)
name = 'OpenCV Python Flip image Vertically and Horizontally'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```
