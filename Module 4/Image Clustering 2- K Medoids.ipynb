# IMAGE CLUSTERING USING K-MEDOIDS ALGORITHM

## Installing the KMedoids(Clustering) library from sklearn.

### Required computational and visualization library. 

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

#Machine learning library
from sklearn_extra.cluster import KMedoids

import warnings
warnings.filterwarnings('ignore')
```

## Create dummy image data

```python
# dummy image data
data = np.array([[7,8], [9,10], [11,5], [4,9], [7,5] ,[2,3], [4,5]])
```

```python
#Visualizing the dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(x = data[:, 0], y = data[:, 1], marker='o', s=100)
plt.title("Dummy Image Data Points")
plt.show()
```

## Initializing Centres Manually

```python
centers = [[4,5], [9.10]]
```

## Creating  a KMedoids model

```python
Kmedoids = KMedoids(n_clusters=2)
Kmedoids.fit(data)

# Assigning the labels
labels = Kmedoids.labels_
```

## Visualizing the cluster formed

```python
# Extraction of unique labels
unique_labels = set(labels)

#Color Codes
color_plot = [
    plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))
]

# Set data points of feature X and feature Y
xy = data

# Plotting feature X and feature Y for each cluster label
plt.figure(figsize=(10, 8))
sns.scatterplot(x=xy[:, 0], y=xy[:, 1], marker='+', hue=labels, s=100)

# Annotate the Centroids using plt.scatter()
plt.scatter(Kmedoids.cluster_centers_[:, 0], Kmedoids.cluster_centers_[:, 1], marker='o', color='red',
            label='Centroids', s=100)

plt.title("KMedoids Clustering on Dummy Data", fontsize=14)
plt.legend()
plt.show()
```
