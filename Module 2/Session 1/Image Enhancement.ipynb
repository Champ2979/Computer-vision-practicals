## Importing necessary libraries

```python
import random  # to generate random integer values
import cv2
from matplotlib import pyplot as plt
```

## Function Definition 

### Adding add_noise() for adding random black and white pixels
#### This function will create a sprinkled salt and pepper effect on the picture. 

```python
def add_noise(img):
    row , col = img.shape
    number_of_pixels = random.randint(300, 1000)
    for i in range (number_of_pixels):
        ycord=random.randint(0, row -1)
        xcord= random.randint(0, col -1)
        img[ycord][xcord] = 255
    number_of_pixels = random.randint(300, 10000)
    for i in range (number_of_pixels):
        ycord = random.randint(0, row -1)
        xcord = random.randint(0, col -1)
        img[ycord][xcord] = 0
    return img
    
```

## Code to read the image.
### The following code will read the image and display it.

```python
image = cv2.imread("apple.jpg")
plt.imshow(image)
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

## Code to create salt and pepper effect on the image.

### The following code fragment uses the add_noise() function to create salt and pepper effect on the image. 

```python
img = cv2.imread("apple.jpg", cv2.IMREAD_GRAYSCALE)
cv2.imwrite('applewithnoise.jpg' ,add_noise(img))
noise_img = cv2.imread('applewithnoise.jpg')
plt.imshow(noise_img)
plt.show()
```

## Enhancing the image

```python
median = cv2.medianBlur(noise_img, 5)
plt.imshow(median)
plt.show()
```
